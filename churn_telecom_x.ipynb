# üì¶ Etapa 6: Expans√£o da Coluna Charges


A coluna `Charges` armazenava dois valores financeiros: o valor mensal e o valor total gasto pelo cliente. Para facilitar a an√°lise, separamos essas informa√ß√µes em duas colunas distintas: `valor_mensal` e `valor_total`. Tamb√©m realizamos a convers√£o da coluna `valor_total` para tipo num√©rico.


````python
# Verifique as colunas dentro da coluna 'account'
print(df_raw['account'].apply(pd.Series).columns)
````

````python
# URL do arquivo JSON
url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json"

# L√™ o JSON
df_raw = pd.read_json(url)

# 1. Expande as colunas aninhadas exceto 'Charges'
df_expandido = pd.concat([
    df_raw.drop(columns=['customer', 'phone', 'internet', 'account']),
    df_raw['customer'].apply(pd.Series),
    df_raw['phone'].apply(pd.Series),
    df_raw['internet'].apply(pd.Series),
    df_raw['account'].apply(pd.Series)
], axis=1)

# 2. Expande a coluna 'Charges' dentro da coluna 'account'
df_expandido = pd.concat([
    df_expandido.drop(columns=['Charges']),
    df_expandido['Charges'].apply(pd.Series)
], axis=1)

# 3. Renomeia as colunas de valores mensais e totais
df_expandido = df_expandido.rename(columns={
    'Monthly': 'valor_mensal',
    'Total': 'valor_total'
})

# 4. Converte 'valor_mensal' e 'valor_total' para float, tratando erros e NaNs
df_expandido['valor_mensal'] = pd.to_numeric(df_expandido['valor_mensal'], errors='coerce').fillna(0)
df_expandido['valor_total'] = pd.to_numeric(df_expandido['valor_total'], errors='coerce').fillna(0)

# 5. Renomeia coluna de ID
df_expandido = df_expandido.rename(columns={"customerID": "id_cliente"})

# 6. Exibe as 5 primeiras linhas como DataFrame
df_expandido.head()
````

# üöÄ Etapa 7: An√°lise Explorat√≥ria Visual do Churn da Telecom X


üìå O que fizemos aqui:

* Filtramos s√≥ as linhas com valores v√°lidos em Churn ('Yes' e 'No')

* Criamos um gr√°fico interativo que mostra a quantidade de clientes que cancelaram e os que ficaram

* Usamos cores diferenciadas para facilitar a visualiza√ß√£o r√°pida

````python
import plotly.express as px

def plot_churn_distribution(df, min_threshold=0.15):
    """
    Gera um gr√°fico de barras da distribui√ß√£o de clientes que cancelaram (Churn).

    Par√¢metros:
    - df (pd.DataFrame): DataFrame que cont√©m a coluna 'Churn' com valores 'Yes' ou 'No'.
    - min_threshold (float): Valor percentual m√≠nimo de churn (entre 0 e 1) para tra√ßar linha de refer√™ncia.

    Retorna:
    - fig (plotly.graph_objects.Figure): Figura do gr√°fico gerado.
    """

    # C√≥pia dos dados para evitar altera√ß√£o do original
    df_churn = df.copy()

    # Garantir que a coluna 'Churn' contenha apenas valores v√°lidos
    df_churn = df_churn[df_churn['Churn'].isin(['Yes', 'No'])].copy()

    # Contagem e percentual de churn
    churn_counts = df_churn['Churn'].value_counts()
    total = churn_counts.sum()
    percent_yes = (churn_counts['Yes'] / total) * 100
    percent_no = (churn_counts['No'] / total) * 100

    # Criar o gr√°fico
    fig = px.histogram(
        df_churn,
        x='Churn',
        color='Churn',
        title='üìä Distribui√ß√£o de Cancelamento (Churn)',
        labels={'Churn': 'Cliente Cancelou?', 'count': 'N√∫mero de Clientes'},
        color_discrete_map={'Yes': 'crimson', 'No': 'mediumseagreen'},
        text_auto=True
    )

    # Adicionar linha de refer√™ncia de limite m√≠nimo de churn
    min_churn_count = total * min_threshold
    fig.add_hline(
        y=min_churn_count,
        line_dash="dash",
        line_color="red",
        annotation_text=f"Limite M√≠nimo de Churn: {min_threshold * 100:.0f}%",
        annotation_position="top left",
        annotation_font_color="red"
    )

    # Anota√ß√µes com percentual
    fig.update_layout(
        xaxis_title='Churn (Cancelamento)',
        yaxis_title='N√∫mero de Clientes',
        showlegend=False,
        font=dict(size=14),
        annotations=[
            dict(
                x='Yes',
                y=churn_counts['Yes'],
                text=f"{percent_yes:.1f}%",
                showarrow=False,
                font=dict(color='crimson', size=16)
            ),
            dict(
                x='No',
                y=churn_counts['No'],
                text=f"{percent_no:.1f}%",
                showarrow=False,
                font=dict(color='mediumseagreen', size=16)
            )
        ]
    )

    return fig

# Gerar o gr√°fico de distribui√ß√£o de Churn
fig_churn_distribution = plot_churn_distribution(df_expandido)

# Exibir o gr√°fico
fig_churn_distribution.show()



````

# üìä Etapa 16: Gr√°fico ‚Äî Taxa de Churn por Tipo de Contrato


Esse gr√°fico nos fornecer√° uma vis√£o de quais os poss√≠veis contratos tem maior ou menor churn (cancelamento)

````python
# Primeiro, calculamos a taxa de churn por tipo de contrato
taxa_churn_contrato = (
    df_expandido.groupby(['Contract', 'Churn'])
    .size()
    .reset_index(name='Quantidade')
)

# Para calcular a taxa percentual, pegamos total por contrato
total_por_contrato = taxa_churn_contrato.groupby('Contract')['Quantidade'].transform('sum')
taxa_churn_contrato['Percentual'] = taxa_churn_contrato['Quantidade'] / total_por_contrato * 100

# Ajustamos o nome dos contratos para portugu√™s (opcional)
taxa_churn_contrato['Contrato'] = taxa_churn_contrato['Contract'].replace({
    'Month-to-month': 'Mensal',
    'One year': 'Anual (1 ano)',
    'Two year': 'Anual (2 anos)'
})

# Gr√°fico de barras empilhadas com percentual
fig_churn_contrato = px.bar(
    taxa_churn_contrato,
    x='Contrato',
    y='Percentual',
    color='Churn',
    color_discrete_map={'No': 'green', 'Yes': 'red'},
    labels={
        'Contrato': 'Tipo de Contrato',
        'Percentual': 'Percentual (%)',
        'Churn': 'Churn (Cancelamento)'
    },
    title='üìâ Taxa de Churn por Tipo de Contrato na Telecom X',
    text=taxa_churn_contrato['Percentual'].apply(lambda x: f'{x:.1f}%')
)

fig_churn_contrato.update_layout(barmode='stack', yaxis=dict(ticksuffix='%'))
fig_churn_contrato.show()
````

### ‚úçÔ∏è Interpreta√ß√£o do Gr√°fico

Este gr√°fico mostra a **taxa de churn** (cancelamento de servi√ßo) para cada tipo de contrato dos clientes da Telecom X, agrupados em:

- **Mensal** (Month-to-month)
- **Anual (1 ano)** (One year)
- **Anual (2 anos)** (Two year)

As barras representam a propor√ß√£o percentual de clientes que **n√£o cancelaram** (em verde) e que **cancelaram** (em vermelho).

**Insights importantes:**

- Clientes com contrato **mensal** apresentam a **maior taxa de churn**, indicando maior propens√£o a cancelar.
- Contratos de 1 ou 2 anos t√™m uma taxa de churn menor, sugerindo maior fideliza√ß√£o.
- Recomenda-se focar estrat√©gias para converter clientes mensais em contratos anuais para reduzir o churn.

Neste notebook, vamos construir um modelo preditivo para identificar clientes que ir√£o cancelar um servi√ßo (churn) utilizando Random Forest. O processo inclui as seguintes etapas principais:

1. **Prepara√ß√£o dos Dados**  
   - Separa√ß√£o das vari√°veis independentes (`features`) e da vari√°vel alvo (`target`), que indica se o cliente deu churn ou n√£o.  
   - Identifica√ß√£o das colunas num√©ricas e categ√≥ricas para aplicar o pr√©-processamento adequado.

2. **Pr√©-processamento**  
   - Aplica√ß√£o de One-Hot Encoding para transformar as vari√°veis categ√≥ricas em um formato num√©rico adequado para o modelo.  
   - Manuten√ß√£o das vari√°veis num√©ricas originais.

3. **Divis√£o Treino/Teste**  
   - Dividimos os dados em conjuntos de treino e teste, preservando a propor√ß√£o da vari√°vel alvo (`stratify=y`) para garantir representatividade.

4. **Balanceamento com SMOTE**  
   - O conjunto de treino √© balanceado utilizando o SMOTE, t√©cnica que gera amostras sint√©ticas para a classe minorit√°ria, prevenindo vi√©s do modelo.

5. **Treinamento e Otimiza√ß√£o do Modelo**  
   - Treinamos um classificador Random Forest.  
   - Utilizamos `GridSearchCV` para encontrar os melhores hiperpar√¢metros do modelo, avaliando por meio de valida√ß√£o cruzada (5 folds) e otimizando a m√©trica F1-score.

6. **Avalia√ß√£o Final**  
   - O modelo otimizado √© avaliado no conjunto de teste.  
   - S√£o calculadas as m√©tricas de acur√°cia, precis√£o, recall, F1-score, al√©m da matriz de confus√£o para an√°lise detalhada do desempenho.

Esta abordagem √© robusta para lidar com dados desbalanceados e combina pr√©-processamento cuidadoso com tuning autom√°tico de hiperpar√¢metros para alcan√ßar alta performance na tarefa de classifica√ß√£o.

---


````python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import joblib

# Carregar os dados do arquivo CSV
df = pd.read_csv('/content/df_expandido.csv')

# Nome da coluna target
target = 'Churn'

# Ver valores √∫nicos da coluna target
print("Valores √∫nicos em Churn:", df[target].unique())

# Mapear valores texto para 0 e 1
map_churn = {'No': 0, 'Yes': 1}
df[target] = df[target].map(map_churn)

# Remover linhas com target nulo
df = df.dropna(subset=[target])

# Garantir que o target seja int
y = df[target].astype(int)

# Definir features
features = df.drop(columns=[target]).columns.tolist()
X = df[features]

# Separar vari√°veis num√©ricas e categ√≥ricas
num_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_features = X.select_dtypes(include=['object', 'bool']).columns.tolist()

# Pipeline de pr√©-processamento
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
    ],
    remainder='passthrough'
)

# Separar dados treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Aplicar pr√©-processamento
X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep = preprocessor.transform(X_test)

# Aplicar SMOTE para balancear as classes
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_prep, y_train)

# Configurar Random Forest
rf = RandomForestClassifier(random_state=42)

# Par√¢metros para GridSearch
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5]
}

# GridSearch com 5 folds e m√©trica F1
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='f1', n_jobs=-1)

# Treinar o modelo com dados balanceados
grid_search.fit(X_train_smote, y_train_smote)

print("Melhores par√¢metros encontrados:", grid_search.best_params_)

# Avaliar no conjunto de teste
y_pred = grid_search.predict(X_test_prep)

print("Acur√°cia:", accuracy_score(y_test, y_pred))
print("\nRelat√≥rio de Classifica√ß√£o:\n", classification_report(y_test, y_pred))
print("Matriz de Confus√£o:\n", confusion_matrix(y_test, y_pred))

# --------- SALVAR MODELO E PREPROCESSOR ---------

# Melhor modelo treinado
modelo_rf = grid_search.best_estimator_

# Salvar modelo e pr√©-processador
joblib.dump(modelo_rf, 'modelo_rf.joblib')
joblib.dump(preprocessor, 'preprocessor.joblib')

print("Arquivos modelo_rf.joblib e preprocessor.joblib salvos com sucesso!")

# --- Download direto no Colab ---
try:
    from google.colab import files
    files.download('modelo_rf.joblib')
    files.download('preprocessor.joblib')
except ImportError:
    print("N√£o est√° rodando no Colab, pule o download autom√°tico.")
````

---

## Passos realizados:

### 1. C√°lculo das probabilidades preditas

### 4. Teste pr√°tico de diferentes thresholds

- Avaliamos o desempenho do modelo para thresholds variados (0.3, 0.4, 0.5, 0.6, 0.7).

### 5. Avalia√ß√£o final com threshold ajustado

> **Recomenda√ß√£o:** Sempre combine an√°lise quantitativa com conhecimento do neg√≥cio para definir o threshold √≥timo.


````python
from sklearn.metrics import (
    roc_auc_score, roc_curve,
    precision_recall_curve, average_precision_score,
    classification_report, confusion_matrix
)

# Supondo que voc√™ j√° tem o grid_search treinado e X_test_prep, y_test definidos

# 1. Obter probabilidades preditas para a classe positiva
y_prob = grid_search.predict_proba(X_test_prep)[:, 1]

# 2. Definir fun√ß√£o para avaliar o modelo para diferentes thresholds
def avaliar_threshold(threshold, y_prob, y_true):
    y_pred = (y_prob >= threshold).astype(int)
    print(f"\n=== Avalia√ß√£o para threshold = {threshold:.2f} ===")
    print("Relat√≥rio de Classifica√ß√£o:")
    print(classification_report(y_true, y_pred))
    print("Matriz de Confus√£o:")
    print(confusion_matrix(y_true, y_pred))
    acc = (y_pred == y_true).mean()
    print(f"Acur√°cia: {acc:.4f}")
    return y_pred

# 3. Plotar curva ROC e Precision-Recall
fpr, tpr, thresholds_roc = roc_curve(y_test, y_prob)
precision, recall, thresholds_pr = precision_recall_curve(y_test, y_prob)
auc_roc = roc_auc_score(y_test, y_prob)
ap_score = average_precision_score(y_test, y_prob)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(fpr, tpr, label=f'AUC-ROC = {auc_roc:.2f}')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('Falso Positivo')
plt.ylabel('Verdadeiro Positivo')
plt.title('Curva ROC')
plt.legend()

plt.subplot(1,2,2)
plt.plot(recall, precision, label=f'AP = {ap_score:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Curva Precision-Recall')
plt.legend()

plt.tight_layout()
plt.show()

# 4. Testar diferentes thresholds para achar um bom ponto de corte
thresholds_teste = [0.3, 0.4, 0.5, 0.6, 0.7]
for th in thresholds_teste:
    avaliar_threshold(th, y_prob, y_test)

# Voc√™ pode escolher o threshold que melhor balanceia recall e precision para seu objetivo
# Por exemplo, suponha que 0.4 √© um bom valor:
threshold_escolhido = 0.4
y_pred_ajustado = (y_prob >= threshold_escolhido).astype(int)

print(f"\n=== Avalia√ß√£o final com threshold ajustado = {threshold_escolhido} ===")
print(classification_report(y_test, y_pred_ajustado))
print("Matriz de Confus√£o:")
print(confusion_matrix(y_test, y_pred_ajustado))
````

def avaliar_threshold(modelo, X_teste, y_teste, threshold=0.5):

    # 2. Aplica√ß√£o do threshold customizado para gerar as predi√ß√µes bin√°rias
    preds = (probs >= threshold).astype(int)

    print("Acur√°cia:", accuracy_score(y_teste, preds))
    print("\n")

    # 4. Retorno das predi√ß√µes bin√°rias geradas
    return preds

## üìà Etapa 28: Predi√ß√£o e Avalia√ß√£o com Threshold Customizado e Salvamento em CSV


Este conjunto de fun√ß√µes permite realizar predi√ß√µes usando um threshold customizado para converter probabilidades em classes bin√°rias, avaliar as predi√ß√µes e, opcionalmente, salvar os resultados em arquivo CSV para posterior an√°lise.

---

## Fun√ß√£o `predizer_com_threshold`

### Descri√ß√£o
Realiza a predi√ß√£o de classes com um limiar (threshold) ajust√°vel, retornando as predi√ß√µes bin√°rias e as probabilidades previstas. Tamb√©m oferece a op√ß√£o de salvar as predi√ß√µes e probabilidades em um arquivo CSV.

### Par√¢metros

## üîç Etapa 30: Avalia√ß√£o de M√©tricas em V√°rios Thresholds e Visualiza√ß√£o Gr√°fica

Esta fun√ß√£o permite avaliar as principais m√©tricas de classifica√ß√£o para diferentes valores de threshold aplicados √†s probabilidades previstas pelo modelo. Os resultados s√£o armazenados em um DataFrame, salvos em arquivo CSV e apresentados graficamente para facilitar a an√°lise do impacto do threshold nas m√©tricas.

---

## Fun√ß√£o `avaliar_varios_thresholds`

### Objetivo
Avaliar m√©tricas de desempenho (Precision, Recall, F1 Score e Acur√°cia) do modelo para uma s√©rie de valores de threshold, possibilitando a escolha do melhor limiar para o problema em quest√£o.

### Par√¢metros
- `modelo`: modelo treinado que possua o m√©todo `predict_proba`.
- `X_test`: features de teste pr√©-processadas.
- `y_test`: r√≥tulos verdadeiros correspondentes a `X_test`.
- `thresholds`: array ou lista de valores float representando os thresholds a serem avaliados (padr√£o: valores de 0.1 a 0.9 com passo 0.1).

### Retorno
- `df_resultados`: DataFrame contendo as m√©tricas calculadas para cada threshold.

---

## Funcionamento
1. Obt√©m as probabilidades previstas da classe positiva para todas as amostras de teste.
2. Para cada threshold especificado:
   - Converte as probabilidades em predi√ß√µes bin√°rias.
   - Calcula as m√©tricas Precision, Recall, F1 Score e Acur√°cia.
   - Armazena os resultados em uma lista de dicion√°rios.
3. Cria um DataFrame a partir dos resultados.
4. Salva o DataFrame em um arquivo CSV chamado `'avaliacao_thresholds.csv'`.
5. Plota as m√©tricas em fun√ß√£o do threshold para visualiza√ß√£o clara das tend√™ncias.
6. Retorna o DataFrame com os resultados para uso posterior.

---

## Visualiza√ß√£o

O gr√°fico gerado mostra as curvas de Precision, Recall, F1 Score e Acur√°cia em fun√ß√£o dos diferentes thresholds testados, permitindo identificar visualmente qual threshold oferece o melhor equil√≠brio entre as m√©tricas.

---

## Exemplo de Uso

```python
df_avaliacao = avaliar_varios_thresholds(grid_search.best_estimator_, X_test_prep, y_test)

print(df_avaliacao)


````python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

def avaliar_varios_thresholds(modelo, X_test, y_test, thresholds=np.arange(0.1, 1.0, 0.1)):
    resultados = []

    probs = modelo.predict_proba(X_test)[:,1]

    for thr in thresholds:
        preds = (probs >= thr).astype(int)
        precision = precision_score(y_test, preds)
        recall = recall_score(y_test, preds)
        f1 = f1_score(y_test, preds)
        acc = accuracy_score(y_test, preds)
        resultados.append({
            'threshold': thr,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'accuracy': acc
        })

    df_resultados = pd.DataFrame(resultados)

    # Salvar CSV
    df_resultados.to_csv('avaliacao_thresholds.csv', index=False)
    print("Resultados salvos em 'avaliacao_thresholds.csv'")

    # Plotar
    plt.figure(figsize=(10,6))
    plt.plot(df_resultados['threshold'], df_resultados['precision'], marker='o', label='Precision')
    plt.plot(df_resultados['threshold'], df_resultados['recall'], marker='o', label='Recall')
    plt.plot(df_resultados['threshold'], df_resultados['f1_score'], marker='o', label='F1 Score')
    plt.plot(df_resultados['threshold'], df_resultados['accuracy'], marker='o', label='Accuracy')
    plt.xlabel('Threshold')
    plt.ylabel('Score')
    plt.title('M√©tricas vs Threshold')
    plt.grid(True)
    plt.legend()
    plt.show()

    return df_resultados

# Exemplo de uso:
df_avaliacao = avaliar_varios_thresholds(grid_search.best_estimator_, X_test_prep, y_test)

print(df_avaliacao)
````
