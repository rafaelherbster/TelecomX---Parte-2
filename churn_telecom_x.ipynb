# 📦 Etapa 6: Expansão da Coluna Charges


A coluna `Charges` armazenava dois valores financeiros: o valor mensal e o valor total gasto pelo cliente. Para facilitar a análise, separamos essas informações em duas colunas distintas: `valor_mensal` e `valor_total`. Também realizamos a conversão da coluna `valor_total` para tipo numérico.


````python
# Verifique as colunas dentro da coluna 'account'
print(df_raw['account'].apply(pd.Series).columns)
````

````python
# URL do arquivo JSON
url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json"

# Lê o JSON
df_raw = pd.read_json(url)

# 1. Expande as colunas aninhadas exceto 'Charges'
df_expandido = pd.concat([
    df_raw.drop(columns=['customer', 'phone', 'internet', 'account']),
    df_raw['customer'].apply(pd.Series),
    df_raw['phone'].apply(pd.Series),
    df_raw['internet'].apply(pd.Series),
    df_raw['account'].apply(pd.Series)
], axis=1)

# 2. Expande a coluna 'Charges' dentro da coluna 'account'
df_expandido = pd.concat([
    df_expandido.drop(columns=['Charges']),
    df_expandido['Charges'].apply(pd.Series)
], axis=1)

# 3. Renomeia as colunas de valores mensais e totais
df_expandido = df_expandido.rename(columns={
    'Monthly': 'valor_mensal',
    'Total': 'valor_total'
})

# 4. Converte 'valor_mensal' e 'valor_total' para float, tratando erros e NaNs
df_expandido['valor_mensal'] = pd.to_numeric(df_expandido['valor_mensal'], errors='coerce').fillna(0)
df_expandido['valor_total'] = pd.to_numeric(df_expandido['valor_total'], errors='coerce').fillna(0)

# 5. Renomeia coluna de ID
df_expandido = df_expandido.rename(columns={"customerID": "id_cliente"})

# 6. Exibe as 5 primeiras linhas como DataFrame
df_expandido.head()
````

<a name="etapa-7-análise-exploratória-visual-do-churn-da-telecom-x"></a>
# 🚀 Etapa 7: Análise Exploratória Visual do Churn da Telecom X


📌 O que fizemos aqui:

* Filtramos só as linhas com valores válidos em Churn ('Yes' e 'No')

* Criamos um gráfico interativo que mostra a quantidade de clientes que cancelaram e os que ficaram

* Usamos cores diferenciadas para facilitar a visualização rápida

````python
import plotly.express as px

def plot_churn_distribution(df, min_threshold=0.15):
    """
    Gera um gráfico de barras da distribuição de clientes que cancelaram (Churn).

    Parâmetros:
    - df (pd.DataFrame): DataFrame que contém a coluna 'Churn' com valores 'Yes' ou 'No'.
    - min_threshold (float): Valor percentual mínimo de churn (entre 0 e 1) para traçar linha de referência.

    Retorna:
    - fig (plotly.graph_objects.Figure): Figura do gráfico gerado.
    """

    # Cópia dos dados para evitar alteração do original
    df_churn = df.copy()

    # Garantir que a coluna 'Churn' contenha apenas valores válidos
    df_churn = df_churn[df_churn['Churn'].isin(['Yes', 'No'])].copy()

    # Contagem e percentual de churn
    churn_counts = df_churn['Churn'].value_counts()
    total = churn_counts.sum()
    percent_yes = (churn_counts['Yes'] / total) * 100
    percent_no = (churn_counts['No'] / total) * 100

    # Criar o gráfico
    fig = px.histogram(
        df_churn,
        x='Churn',
        color='Churn',
        title='📊 Distribuição de Cancelamento (Churn)',
        labels={'Churn': 'Cliente Cancelou?', 'count': 'Número de Clientes'},
        color_discrete_map={'Yes': 'crimson', 'No': 'mediumseagreen'},
        text_auto=True
    )

    # Adicionar linha de referência de limite mínimo de churn
    min_churn_count = total * min_threshold
    fig.add_hline(
        y=min_churn_count,
        line_dash="dash",
        line_color="red",
        annotation_text=f"Limite Mínimo de Churn: {min_threshold * 100:.0f}%",
        annotation_position="top left",
        annotation_font_color="red"
    )

    # Anotações com percentual
    fig.update_layout(
        xaxis_title='Churn (Cancelamento)',
        yaxis_title='Número de Clientes',
        showlegend=False,
        font=dict(size=14),
        annotations=[
            dict(
                x='Yes',
                y=churn_counts['Yes'],
                text=f"{percent_yes:.1f}%",
                showarrow=False,
                font=dict(color='crimson', size=16)
            ),
            dict(
                x='No',
                y=churn_counts['No'],
                text=f"{percent_no:.1f}%",
                showarrow=False,
                font=dict(color='mediumseagreen', size=16)
            )
        ]
    )

    return fig

# Gerar o gráfico de distribuição de Churn
fig_churn_distribution = plot_churn_distribution(df_expandido)

# Exibir o gráfico
fig_churn_distribution.show()



````

## 🎯 Etapa 22: Modelo de Classificação com Random Forest e Balanceamento de Classes usando SMOTE


Neste notebook, vamos construir um modelo preditivo para identificar clientes que irão cancelar um serviço (churn) utilizando Random Forest. O processo inclui as seguintes etapas principais:

1. **Preparação dos Dados**  
   - Separação das variáveis independentes (`features`) e da variável alvo (`target`), que indica se o cliente deu churn ou não.  
   - Identificação das colunas numéricas e categóricas para aplicar o pré-processamento adequado.

2. **Pré-processamento**  
   - Aplicação de One-Hot Encoding para transformar as variáveis categóricas em um formato numérico adequado para o modelo.  
   - Manutenção das variáveis numéricas originais.

3. **Divisão Treino/Teste**  
   - Dividimos os dados em conjuntos de treino e teste, preservando a proporção da variável alvo (`stratify=y`) para garantir representatividade.

4. **Balanceamento com SMOTE**  
   - O conjunto de treino é balanceado utilizando o SMOTE, técnica que gera amostras sintéticas para a classe minoritária, prevenindo viés do modelo.

5. **Treinamento e Otimização do Modelo**  
   - Treinamos um classificador Random Forest.  
   - Utilizamos `GridSearchCV` para encontrar os melhores hiperparâmetros do modelo, avaliando por meio de validação cruzada (5 folds) e otimizando a métrica F1-score.

6. **Avaliação Final**  
   - O modelo otimizado é avaliado no conjunto de teste.  
   - São calculadas as métricas de acurácia, precisão, recall, F1-score, além da matriz de confusão para análise detalhada do desempenho.

Esta abordagem é robusta para lidar com dados desbalanceados e combina pré-processamento cuidadoso com tuning automático de hiperparâmetros para alcançar alta performance na tarefa de classificação.

---


````python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import joblib

# Carregar os dados do arquivo CSV
df = pd.read_csv('/content/df_expandido.csv')

# Nome da coluna target
target = 'Churn'

# Ver valores únicos da coluna target
print("Valores únicos em Churn:", df[target].unique())

# Mapear valores texto para 0 e 1
map_churn = {'No': 0, 'Yes': 1}
df[target] = df[target].map(map_churn)

# Remover linhas com target nulo
df = df.dropna(subset=[target])

# Garantir que o target seja int
y = df[target].astype(int)

# Definir features
features = df.drop(columns=[target]).columns.tolist()
X = df[features]

# Separar variáveis numéricas e categóricas
num_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_features = X.select_dtypes(include=['object', 'bool']).columns.tolist()

# Pipeline de pré-processamento
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
    ],
    remainder='passthrough'
)

# Separar dados treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Aplicar pré-processamento
X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep = preprocessor.transform(X_test)

# Aplicar SMOTE para balancear as classes
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_prep, y_train)

# Configurar Random Forest
rf = RandomForestClassifier(random_state=42)

# Parâmetros para GridSearch
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5]
}

# GridSearch com 5 folds e métrica F1
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='f1', n_jobs=-1)

# Treinar o modelo com dados balanceados
grid_search.fit(X_train_smote, y_train_smote)

print("Melhores parâmetros encontrados:", grid_search.best_params_)

# Avaliar no conjunto de teste
y_pred = grid_search.predict(X_test_prep)

print("Acurácia:", accuracy_score(y_test, y_pred))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred))
print("Matriz de Confusão:\n", confusion_matrix(y_test, y_pred))

# --------- SALVAR MODELO E PREPROCESSOR ---------

# Melhor modelo treinado
modelo_rf = grid_search.best_estimator_

# Salvar modelo e pré-processador
joblib.dump(modelo_rf, 'modelo_rf.joblib')
joblib.dump(preprocessor, 'preprocessor.joblib')

print("Arquivos modelo_rf.joblib e preprocessor.joblib salvos com sucesso!")

# --- Download direto no Colab ---
try:
    from google.colab import files
    files.download('modelo_rf.joblib')
    files.download('preprocessor.joblib')
except ImportError:
    print("Não está rodando no Colab, pule o download automático.")
````

## 📊 Etapa 25: Avaliação do Modelo com Ajuste de Threshold e Curvas ROC / Precision-Recall


Neste trecho de código, aprofundamos a avaliação do modelo de classificação balanceado pelo SMOTE e otimizado via Random Forest, focando em:

- **Análise das probabilidades preditas** para a classe positiva;
- **Visualização das curvas ROC e Precision-Recall**, métricas essenciais para avaliar classificadores binários, especialmente em datasets desbalanceados;
- **Ajuste do threshold de decisão** para encontrar o ponto ótimo que balanceia recall e precisão, indo além do valor padrão (0.5).

---

## Passos realizados:

### 1. Cálculo das probabilidades preditas

- Extraímos as probabilidades previstas para a classe positiva (`y_prob`) usando o método `predict_proba` do melhor modelo encontrado (`grid_search`).

### 2. Função para avaliação do modelo em diferentes thresholds

- Criamos uma função que, dado um valor de threshold, converte probabilidades em predições binárias, calcula e imprime:
  - Relatório de classificação detalhado (precisão, recall, F1-score);
  - Matriz de confusão;
  - Acurácia.

### 3. Plotagem das curvas ROC e Precision-Recall

- **Curva ROC:** mostra a relação entre taxa de falso positivo (FPR) e verdadeiro positivo (TPR) para vários thresholds.
- **Curva Precision-Recall:** essencial para avaliar modelos em cenários desbalanceados, destacando o trade-off entre precisão e recall.
- Calculamos e exibimos as métricas AUC-ROC e Average Precision (AP).

### 4. Teste prático de diferentes thresholds

- Avaliamos o desempenho do modelo para thresholds variados (0.3, 0.4, 0.5, 0.6, 0.7).

### 5. Avaliação final com threshold ajustado

- Aplicamos o threshold escolhido (exemplo: 0.4) para obter a predição final e avaliamos o desempenho correspondente.

---

## Importância do ajuste de threshold

- O threshold padrão de 0.5 pode não ser ideal, especialmente em casos de classes desbalanceadas.
- Ajustar o threshold pode melhorar significativamente o desempenho do modelo para objetivos específicos, como minimizar falsos negativos em problemas críticos (ex.: churn, fraude, doenças).
- As curvas e métricas auxiliam na escolha informada do threshold mais adequado.

---

> **Recomendação:** Sempre combine análise quantitativa com conhecimento do negócio para definir o threshold ótimo.


````python
from sklearn.metrics import (
    roc_auc_score, roc_curve,
    precision_recall_curve, average_precision_score,
    classification_report, confusion_matrix
)

# Supondo que você já tem o grid_search treinado e X_test_prep, y_test definidos

# 1. Obter probabilidades preditas para a classe positiva
y_prob = grid_search.predict_proba(X_test_prep)[:, 1]

# 2. Definir função para avaliar o modelo para diferentes thresholds
def avaliar_threshold(threshold, y_prob, y_true):
    y_pred = (y_prob >= threshold).astype(int)
    print(f"\n=== Avaliação para threshold = {threshold:.2f} ===")
    print("Relatório de Classificação:")
    print(classification_report(y_true, y_pred))
    print("Matriz de Confusão:")
    print(confusion_matrix(y_true, y_pred))
    acc = (y_pred == y_true).mean()
    print(f"Acurácia: {acc:.4f}")
    return y_pred

# 3. Plotar curva ROC e Precision-Recall
fpr, tpr, thresholds_roc = roc_curve(y_test, y_prob)
precision, recall, thresholds_pr = precision_recall_curve(y_test, y_prob)
auc_roc = roc_auc_score(y_test, y_prob)
ap_score = average_precision_score(y_test, y_prob)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(fpr, tpr, label=f'AUC-ROC = {auc_roc:.2f}')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('Falso Positivo')
plt.ylabel('Verdadeiro Positivo')
plt.title('Curva ROC')
plt.legend()

plt.subplot(1,2,2)
plt.plot(recall, precision, label=f'AP = {ap_score:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Curva Precision-Recall')
plt.legend()

plt.tight_layout()
plt.show()

# 4. Testar diferentes thresholds para achar um bom ponto de corte
thresholds_teste = [0.3, 0.4, 0.5, 0.6, 0.7]
for th in thresholds_teste:
    avaliar_threshold(th, y_prob, y_test)

# Você pode escolher o threshold que melhor balanceia recall e precision para seu objetivo
# Por exemplo, suponha que 0.4 é um bom valor:
threshold_escolhido = 0.4
y_pred_ajustado = (y_prob >= threshold_escolhido).astype(int)

print(f"\n=== Avaliação final com threshold ajustado = {threshold_escolhido} ===")
print(classification_report(y_test, y_pred_ajustado))
print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred_ajustado))
````

````python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

def predizer_com_threshold(modelo, X, threshold=0.5, salvar_csv=False, nome_arquivo='predicoes.csv'):
    """
    Realiza predição com threshold customizado, avalia e salva resultados.

    Parâmetros:
    - modelo: modelo treinado com método predict_proba
    - X: features pré-processadas para predição
    - threshold: float, limiar para converter probabilidades em classes
    - salvar_csv: bool, se True salva as predições em CSV
    - nome_arquivo: str, nome do arquivo CSV para salvar predições

    Retorna:
    - preds: array com predições binárias
    """
    # Pega probabilidades da classe positiva
    probs = modelo.predict_proba(X)[:, 1]

    # Aplica threshold customizado
    preds = (probs >= threshold).astype(int)

    # Avaliação só se houver rótulos verdadeiros (opcional)
    # Aqui assumimos que o usuário fornece X já separado para teste e tem y_test separado fora desta função
    # Caso queira passar y_test, podemos alterar a função

    if salvar_csv:
        df_result = pd.DataFrame({'probabilidade': probs, 'predicao': preds})
        df_result.to_csv(nome_arquivo, index=False)
        print(f'Predições salvas em {nome_arquivo}')

    return preds, probs

def avaliar_predicoes(y_true, preds):
    """
    Avalia as predições geradas e imprime relatório.

    Parâmetros:
    - y_true: array com valores verdadeiros
    - preds: array com predições binárias
    """
    print("Relatório de Classificação:")
    print(classification_report(y_true, preds))
    print("Matriz de Confusão:")
    print(confusion_matrix(y_true, preds))
    print("Acurácia:", accuracy_score(y_true, preds))

# Uso:

# Define threshold
threshold_escolhido = 0.4

# Prediz com threshold customizado
y_pred_custom, y_prob_custom = predizer_com_threshold(grid_search.best_estimator_, X_test_prep, threshold=threshold_escolhido, salvar_csv=True)

# Avalia com os dados verdadeiros
avaliar_predicoes(y_test, y_pred_custom)

import os

nome_arquivo = 'predicoes.csv'
print("Arquivo será salvo em:", os.path.abspath(nome_arquivo))

````

## 🔍 Etapa 30: Avaliação de Métricas em Vários Thresholds e Visualização Gráfica

Esta função permite avaliar as principais métricas de classificação para diferentes valores de threshold aplicados às probabilidades previstas pelo modelo. Os resultados são armazenados em um DataFrame, salvos em arquivo CSV e apresentados graficamente para facilitar a análise do impacto do threshold nas métricas.

---

## Função `avaliar_varios_thresholds`

### Objetivo
Avaliar métricas de desempenho (Precision, Recall, F1 Score e Acurácia) do modelo para uma série de valores de threshold, possibilitando a escolha do melhor limiar para o problema em questão.

### Parâmetros
- `modelo`: modelo treinado que possua o método `predict_proba`.
- `X_test`: features de teste pré-processadas.
- `y_test`: rótulos verdadeiros correspondentes a `X_test`.
- `thresholds`: array ou lista de valores float representando os thresholds a serem avaliados (padrão: valores de 0.1 a 0.9 com passo 0.1).

### Retorno
- `df_resultados`: DataFrame contendo as métricas calculadas para cada threshold.

---

## Funcionamento
1. Obtém as probabilidades previstas da classe positiva para todas as amostras de teste.
2. Para cada threshold especificado:
   - Converte as probabilidades em predições binárias.
   - Calcula as métricas Precision, Recall, F1 Score e Acurácia.
   - Armazena os resultados em uma lista de dicionários.
3. Cria um DataFrame a partir dos resultados.
4. Salva o DataFrame em um arquivo CSV chamado `'avaliacao_thresholds.csv'`.
5. Plota as métricas em função do threshold para visualização clara das tendências.
6. Retorna o DataFrame com os resultados para uso posterior.

---

## Visualização

O gráfico gerado mostra as curvas de Precision, Recall, F1 Score e Acurácia em função dos diferentes thresholds testados, permitindo identificar visualmente qual threshold oferece o melhor equilíbrio entre as métricas.

---

## Exemplo de Uso

```python
df_avaliacao = avaliar_varios_thresholds(grid_search.best_estimator_, X_test_prep, y_test)

print(df_avaliacao)


````python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

def avaliar_varios_thresholds(modelo, X_test, y_test, thresholds=np.arange(0.1, 1.0, 0.1)):
    resultados = []

    probs = modelo.predict_proba(X_test)[:,1]

    for thr in thresholds:
        preds = (probs >= thr).astype(int)
        precision = precision_score(y_test, preds)
        recall = recall_score(y_test, preds)
        f1 = f1_score(y_test, preds)
        acc = accuracy_score(y_test, preds)
        resultados.append({
            'threshold': thr,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'accuracy': acc
        })

    df_resultados = pd.DataFrame(resultados)

    # Salvar CSV
    df_resultados.to_csv('avaliacao_thresholds.csv', index=False)
    print("Resultados salvos em 'avaliacao_thresholds.csv'")

    # Plotar
    plt.figure(figsize=(10,6))
    plt.plot(df_resultados['threshold'], df_resultados['precision'], marker='o', label='Precision')
    plt.plot(df_resultados['threshold'], df_resultados['recall'], marker='o', label='Recall')
    plt.plot(df_resultados['threshold'], df_resultados['f1_score'], marker='o', label='F1 Score')
    plt.plot(df_resultados['threshold'], df_resultados['accuracy'], marker='o', label='Accuracy')
    plt.xlabel('Threshold')
    plt.ylabel('Score')
    plt.title('Métricas vs Threshold')
    plt.grid(True)
    plt.legend()
    plt.show()

    return df_resultados

# Exemplo de uso:
df_avaliacao = avaliar_varios_thresholds(grid_search.best_estimator_, X_test_prep, y_test)

print(df_avaliacao)
````
